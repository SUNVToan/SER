```python
def insert_cabinet(cabinet, to_insert):
    check_location = len(cabinet) - 1
    insert_location = 0
    while check_location >= 0:
        if to_insert > cabinet[check_location]:
            insert_location = check_location + 1
            check_location = -1
    check_location = check_location - 1
    cabinet.insert(insert_location, to_insert)
    return cabinet


cabinet = [1, 2, 3, 3, 4, 6, 8, 12]
newcabinet = insert_cabinet(cabinet, 5)
print(newcabinet)
```

```python
def get_mean_prediction(prediction):
    """
    Tính giá trị trung bình cho mỗi loại cảm xúc trên toàn bộ đoạn âm thanh.

    Args:
      prediction: Danh sách chứa các danh sách con, mỗi danh sách con là
                  dự đoán cho một khung thời gian (7 giá trị tương ứng với 7
                  loại cảm xúc).

    Returns:
      list: Danh sách chứa giá trị trung bình cho mỗi loại cảm xúc.
    """

    mean_prediction = [0] * 7  # Khởi tạo danh sách để lưu trữ giá trị trung bình

    for frame_prediction in prediction:
        # Cộng dồn giá trị cho mỗi loại cảm xúc trong mỗi khung thời gian
        for i, value in enumerate(frame_prediction):
            mean_prediction[i] += value

    # Chia tổng cho số lượng khung thời gian để lấy giá trị trung bình
    for i in range(len(mean_prediction)):
        mean_prediction[i] /= len(prediction)

    return mean_prediction


# Ví dụ sử dụng
prediction = [
    [0.8, 0.1, 0.05, 0.02, 0.01, 0.01, 0.01],
    [0.7, 0.2, 0.08, 0.03, 0.01, 0.01, 0.01],
    [0.6, 0.3, 0.07, 0.04, 0.02, 0.01, 0.01],
]

mean_prediction = get_mean_prediction(prediction)
print(mean_prediction)
```

```python
def linear_search(arr):
    """
    Tìm tất cả chỉ mục của giá trị cao nhất trong danh sách.

    Args:
      arr: Danh sách các giá trị.

    Returns:
      list: Danh sách chứa các chỉ mục của giá trị cao nhất.
    """

    max_value = max(arr)  # Tìm giá trị cao nhất
    max_indices = []  # Khởi tạo danh sách để lưu trữ các chỉ mục

    for i, value in enumerate(arr):
        if value == max_value:
            max_indices.append(i)  # Thêm chỉ mục vào danh sách

    return max_indices
```

```python
def get_highest_prediction(prediction):
    """
    Lấy giá trị cao nhất trên toàn bộ đoạn âm thanh cho mỗi loại cảm xúc.

    Args:
      prediction: Danh sách chứa các danh sách con, mỗi danh sách con là một
                  dự đoán cho một khung thời gian (7 giá trị tương ứng với 7
                  loại cảm xúc).

    Returns:
      dictionary: Từ điển chứa tỷ lệ phần trăm cao nhất cho mỗi loại cảm xúc.
    """

    angry = 0
    disgust = 0
    fear = 0
    happy = 0
    neutral = 0
    sad = 0
    surprise = 0

    for item in prediction:
        angry = max(angry, item[0])
        disgust = max(disgust, item[1])
        fear = max(fear, item[2])
        happy = max(happy, item[3])
        neutral = max(neutral, item[4])
        sad = max(sad, item[5])
        surprise = max(surprise, item[6])

    return [angry, disgust, fear, happy, neutral, sad, surprise]
```

```python
import re
import pandas as pd
from datetime import datetime
from collections import Counter

# Read log file content
log_file_path = "result.log"
with open(log_file_path, "r") as file:
    log_content = file.read()

# Initialize variables to store information
files_info = []
lines = log_content.split("\n")
current_file_info = {}

# Define regex patterns
file_pattern = re.compile(r"File uploaded: (.+\.wav)")
audio_pattern = re.compile(
    r"Audio loaded successfully: (.+), Sample rate: (\d+), Length: (\d+)"
)
mfcc_pattern = re.compile(r"MFCC shape: \((\d+), (\d+), (\d+)\)")
emotion_counts_pattern = re.compile(r"Emotion counts: (Counter\(.+\))")
emotion_detected_pattern = re.compile(r"Emotion detected: (\w+)")
timestamp_pattern = re.compile(r"\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2},\d+")

# Process log lines
for line in lines:
    if file_match := file_pattern.search(line):
        if current_file_info:
            current_file_info["true_label"] = current_file_info["file"].split("_")[1]
            current_file_info["correct_prediction"] = (
                current_file_info["true_label"] == current_file_info["emotion_detected"]
            )
            files_info.append(current_file_info)
        current_file_info = {"file": file_match.group(1)}
        current_file_info["start_time"] = timestamp_pattern.search(line).group()
    elif audio_match := audio_pattern.search(line):
        current_file_info["audio_path"] = audio_match.group(1)
        current_file_info["sample_rate"] = int(audio_match.group(2))
        current_file_info["length"] = int(audio_match.group(3))
    elif mfcc_match := mfcc_pattern.search(line):
        current_file_info["mfcc_shape"] = (
            int(mfcc_match.group(1)),
            int(mfcc_match.group(2)),
            int(mfcc_match.group(3)),
        )
    elif emotion_counts_match := emotion_counts_pattern.search(line):
        current_file_info["emotion_counts"] = eval(emotion_counts_match.group(1))
    elif emotion_detected_match := emotion_detected_pattern.search(line):
        current_file_info["emotion_detected"] = emotion_detected_match.group(1)
        current_file_info["end_time"] = timestamp_pattern.search(line).group()
        start_time = current_file_info["start_time"]
        end_time = current_file_info["end_time"]
        time_format = "%Y-%m-%d %H:%M:%S,%f"
        processing_time = (
            datetime.strptime(end_time, time_format)
            - datetime.strptime(start_time, time_format)
        ).total_seconds()
        current_file_info["processing_time"] = processing_time

# Add the last file's information
if current_file_info:
    current_file_info["true_label"] = current_file_info["file"].split("_")[1]
    current_file_info["correct_prediction"] = (
        current_file_info["true_label"] == current_file_info["emotion_detected"]
    )
    files_info.append(current_file_info)

# Convert to DataFrame
df = pd.DataFrame(files_info)

# Summary of correct predictions
total_samples = len(df)
correct_predictions = df["correct_prediction"].sum()
accuracy = correct_predictions / total_samples

# Save to Excel
output_path = "emotion_recognition_report.xlsx"
df.to_excel(output_path, index=False)

print(f"Excel report saved to: {output_path}")
print(f"Total samples: {total_samples}")
print(f"Correct predictions: {correct_predictions}")
print(f"Accuracy: {accuracy:.2%}")
```

```python
import pandas as pd
import re
from datetime import datetime

# Đọc file log
log_file = "result_final.log"
with open(log_file, "r") as f:
    logs = f.readlines()

# Khởi tạo danh sách để lưu dữ liệu
data = []

# Các regex để trích xuất thông tin
file_pattern = re.compile(r"File uploaded: (.+\.wav)")
start_time_pattern = re.compile(
    r"(\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2},\d{3}) - root - INFO - File uploaded:"
)
audio_pattern = re.compile(
    r"Audio loaded successfully: (.+), Sample rate: (\d+), Length: (\d+)"
)
mfcc_shape_pattern = re.compile(r"MFCC extracted successfully: Shape: \((\d+),\)")
emotion_pattern = re.compile(r"Transformed Prediction: \[\[\'(\w+)\'\]\]")

current_entry = {}
for log in logs:
    if "File uploaded" in log:
        if current_entry:
            end_time = datetime.strptime(log[:23], "%Y-%m-%d %H:%M:%S,%f")
            current_entry["end_time"] = end_time
            current_entry["processing_time"] = (
                end_time - current_entry["start_time"]
            ).total_seconds()
            data.append(current_entry)
            current_entry = {}

        file_match = file_pattern.search(log)
        start_time_match = start_time_pattern.search(log)
        if file_match and start_time_match:
            current_entry["file"] = file_match.group(1)
            current_entry["start_time"] = datetime.strptime(
                start_time_match.group(1), "%Y-%m-%d %H:%M:%S,%f"
            )
            current_entry["true_label"] = (
                file_match.group(1).split("_")[-1].split(".")[0]
            )

    if "Audio loaded successfully" in log:
        audio_match = audio_pattern.search(log)
        if audio_match:
            current_entry["audio_path"] = audio_match.group(1)
            current_entry["sample_rate"] = int(audio_match.group(2))
            current_entry["length"] = int(audio_match.group(3))

    if "MFCC extracted successfully" in log:
        mfcc_shape_match = mfcc_shape_pattern.search(log)
        if mfcc_shape_match:
            current_entry["mfcc_shape"] = (1, int(mfcc_shape_match.group(1)), 1)

    if "Transformed Prediction" in log:
        emotion_match = emotion_pattern.search(log)
        if emotion_match:
            current_entry["emotion_detected"] = emotion_match.group(1)
            current_entry["correct_prediction"] = (
                current_entry["true_label"] == current_entry["emotion_detected"]
            )

# Thêm entry cuối cùng
if current_entry:
    end_time = datetime.strptime(log[:23], "%Y-%m-%d %H:%M:%S,%f")
    current_entry["end_time"] = end_time
    current_entry["processing_time"] = (
        end_time - current_entry["start_time"]
    ).total_seconds()
    data.append(current_entry)

# Tạo dataframe từ danh sách dữ liệu
df = pd.DataFrame(
    data,
    columns=[
        "file",
        "start_time",
        "audio_path",
        "sample_rate",
        "length",
        "mfcc_shape",
        "emotion_detected",
        "end_time",
        "processing_time",
        "true_label",
        "correct_prediction",
    ],
)

# Lưu dataframe dưới dạng file Excel
output_excel_file = "Emotion_Predicted_Output.xlsx"
df.to_excel(output_excel_file, index=False)

print(f"File Excel đã được lưu tại {output_excel_file}")
```
